# K-Means Clustering


The number of clusters **K** is determined by the **means (centriods)** that we will have 

```python
max_k=12 #nb of potential clusters

inertia=[]
k_values=range(2,max_k + 1) # start from 

for k in k_values:
    kmeans= KMeans(n_clusters=k, random_state=42,max_iter=1000) #setup the clusters

    kmeans.fit_predict(scaled_data_df) #predicts the clusters
    inertia.append(kmeans.inertia_) #captures distance

plt.figure(figsize=(14,6))

plt.plot(k_values,inertia,marker='o')
plt.title('KMeans Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)


plt.show()
```

As the graph shows the distance between the different clusters is in a descending order but at **some point it reached stability**


**Which indicates that the distance between the clusters didn't really move a lot**


These are the points __k=4/k=5 when the cluster's centriod didn't shift as much__

## Silhouette Score

$$
s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
$$

-  $s(i)$:is the silhouette score for a single sample $i$
-  $a(i)$: is the average distance between $i$ and all other points in the same cluster
-  $b(i)$:is the minimum average distance between $i$ and all points in the nearest cluster to which $i$ does not belong.


The silhoutte score ranges between $[-1,1]$, a higher value indicates more distinct clusters.

```python
max_k=12 #nb of potential clusters

inertia=[]
silhouette_scores=[]
k_values=range(2,max_k + 1) # start from 

for k in k_values:
    kmeans= KMeans(n_clusters=k, random_state=42,max_iter=1000) #setup the clusters and spread them

    cluster_labels=kmeans.fit_predict(scaled_data_df) #predicts the clusters label->mean centriods
    sil_score= silhouette_score(scaled_data_df,cluster_labels)
    silhouette_scores.append(sil_score)
    inertia.append(kmeans.inertia_) #captures distance

plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(k_values,inertia,marker='o')
plt.title('KMeans Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(k_values,silhouette_scores,marker='o',color='orange')
plt.title('Silhouette for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True)

plt.show()
```

**In the Silhouette Graph** we can clearly visualize whether the clusters are overlapping or the clusters are taking individual form 


In the previous plot the clusters **k=4 and k=5 showcased a stable edge** over other clusters groups indicating that the number 4,5 might be fit

To take that analogy futher, We conducted the silhouette score to showcase how the value of the score is reflecting on the cluster groups, **_for k=4 the silhouette score is  ~0.42 which is higher than that of k=5 about 0.40._**


_So, k=4 will **potentially have less overlap** over it's clusters compared to k=5_


Therefore,we are going to use 4 clusters



```python
kmeans=KMeans(n_clusters=4,random_state=42,max_iter=1000)

cluster_labels=kmeans.fit_predict(scaled_data_df)

cluster_labels
```

# Cluster Analysis

```python

cluster_colors = {
    0: '#1f77b4',
    1: '#ff7f0e',
    2: '#2ca02c',
    3: '#d62728'
}

colors = non_outliers_df['Cluster'].map(cluster_colors)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

scatter = ax.scatter(
    non_outliers_df['MonetaryValue'],
    non_outliers_df['Frequency'],
    non_outliers_df['Recency'],
    c=colors,
    marker='o'
)

# Annotation section 
for cluster_id, color in cluster_colors.items():
    cluster_data = non_outliers_df[non_outliers_df['Cluster'] == cluster_id]
    
    # centroids
    x_mean = cluster_data['MonetaryValue'].mean()
    y_mean = cluster_data['Frequency'].mean()
    z_mean = cluster_data['Recency'].mean()
    
    count = len(cluster_data)
    
    ax.text(
        x_mean, y_mean, z_mean,
        f'Cluster {cluster_id}\n(n={count})',
        color='black',
        fontsize=10,
        ha='center',
        va='center'
    )

ax.set_xlabel('Monetary Value')
ax.set_ylabel('Frequency')
ax.set_zlabel('Recency')
ax.set_title('3D Scatter Plot of Customer Data by Cluster')

plt.show()
```

4. **Cluster 3 (Red): "Reward"**

- __Rationale__: This cluster includes high-value, very frequent buyers, many of whom are still actively purchasing. They are your most loyal customers, and rewarding their loyalty is key to maintaining their engagement.


- __Action__: Implement a robust loyalty program, provide exclusive offers, and recognize their loyalty to keep them engaged and satisfied.
